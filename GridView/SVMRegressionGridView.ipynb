```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
data = pd.read_csv('D:/Ai course/2.MachineLearning/insurance_pre.csv')
data
```

```python
dataset=pd.get_dummies(data,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dependent=dataset[['charges']]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR

param_grid={'kernel':['linear','rbf','poly','sigmoid'],'C':[10,100,1000,2000,3000,10000],
           'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid, refit=True, verbose=3, n_jobs=-1)

grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
grid_prediction=grid.predict(X_test)

from sklearn.metrics import r2_score
r_score=r2_score(Y_test,grid_prediction)
print('The R score values for best parameter{}:'.format(grid.best_params_),r_score)
```
