```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xgboost
```

```python
data= pd.read_csv('CKD.csv')
data
```

```python
dataset=pd.get_dummies(data,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
dependent=dataset['classification_yes']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from xgboost import XGBRFClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'n_estimators':[50,100,1000],'loss':['log_loss', 'deviance', 'exponential'],
           'criterion':['friedman_mse', 'squared_error'],'max_features':['auto', 'sqrt', 'log2']}
grid=GridSearchCV(XGBRFClassifier(),param_grid,refit=True,verbose=0,n_jobs=1,scoring='f1_weighted')
grid.fit(X_train,Y_train)
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(X_test)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, grid_predictions)

from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,grid_predictions)
```

```python
print(cm)
```

```python
print(clf_report)
```

```python
from sklearn.metrics import f1_score
f1_macro=f1_score(Y_test,grid_predictions,average='weighted')
print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print(grid.best_params_)
```

```python
from sklearn.metrics import roc_auc_score

roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
